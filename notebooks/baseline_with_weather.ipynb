```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
import time
from sklearn.metrics import root_mean_squared_error as rmse
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
import pandas as pd
import numpy as np
```

```python
features = ['Year', 'Month', 'CRSDepTime', 'DayofMonth', 'DayOfWeek', 'CRSArrTime', 'dep_mx2t', 'dep_mn2t', 'dep_mxtpr', 'dep_fg10', 'dep_u10', 'dep_v10', 
                 'arr_mx2t', 'arr_mn2t', 'arr_mxtpr', 'arr_fg10', 'arr_u10', 'arr_v10',  'Dest', 'Origin', 'Reporting_Airline', 'Tail_Number']
target = ['DepDelayMinutes']
```

```python
df = pd.read_csv('C:\\Users\\natha\\Documents\\combined_weather_dec2020\\2020_12_with_weather.csv')
df = df.dropna()
```

```python
from sklearn.model_selection import train_test_split
```

```python
traindf, testdf = train_test_split(df, test_size=0.2)
```

```python
ct = ColumnTransformer(
    [
    ('onehot', OneHotEncoder(handle_unknown='ignore'), ['Dest', 'Origin', 'Reporting_Airline', 'Tail_Number'])
    ], remainder = 'passthrough'
)

linear_reg = Pipeline([
    ('columntransform', ct),
    ('model', LinearRegression())
])
```

```python
start_fit_lin = time.perf_counter()
linear_reg.fit(traindf[features], traindf[target])
end_fit_lin = time.perf_counter()
time_fit_lin = end_fit_lin - start_fit_lin
print(f'fit took {time_fit_lin}')

start_pred_lin = time.perf_counter()
lin_pred = linear_reg.predict(testdf[features])
end_pred_lin = time.perf_counter()
time_pred_lin = end_pred_lin - start_pred_lin
print(f'pred took {time_pred_lin}')

print(rmse(testdf[target], lin_pred))
print(mean_absolute_error(testdf[target], lin_pred))
print(r2_score(testdf[target], lin_pred))
```
