```python
# import sys
import os
import math
import json
from typing import List, Dict, Optional, Tuple

import seaborn as sns
import plotly.figure_factory as ff

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import glob
import zipfile
```

```python
path = os.path.join("cleaned_files_dec2020_nov2025", "cleaned_*.csv")
all_files = glob.glob(path)
li = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
import missingno as msno
# Visualize missing points
msno.matrix(li.sample(1000))

# Show correlation of missingness
msno.heatmap(li)

```

```python

#Turn float to int
li["DDMint"] = li["DepDelayMinutes"].astype("Int64")


# Histogram showing Frequency of total time delay
plt.figure(figsize=(7,6))
sns.histplot(li["DDMint"], bins=range(0, li["DDMint"].max()+1), kde=False)
plt.xlim(0, 1500)      # Zooms X-axis to see delays between 0 and 500
plt.ylim(0, 60000)
plt.title("Distribution of Time of Delay per Flight")
plt.xlabel("Time of Delay")
plt.ylabel("Frequency")
plt.show()

# Time series of >15-min-delay per DayOfWeek
monthly_counts = (
    li[li["DepDel15"] == 1].groupby("DayOfWeek").size()
)

ax = monthly_counts.plot(figsize=(7, 6))
ax.set_title("15-min Delays per Day of Week")
ax.set_ylabel("Count")



# Heatmap: >15-min-delay per day of week Ã— hour of day

tmp = li[li["DepDel15"] == 1]
pivot = (
    tmp.groupby(["Month", "DayOfWeek"])
       .size()
       .unstack(fill_value=0)
)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot, cmap="Reds")
plt.title(">15 minute delays by Day of Week and Hour of Day")
plt.xlabel("Day of Week")
plt.ylabel("Month")
plt.tight_layout()
```
